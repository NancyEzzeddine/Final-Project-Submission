```python
import pandas as pd
```

```python
# Load the datasets
file_path_1 = 'Economic_activity_ILO.csv'  
econ_activity = pd.read_csv(file_path_1)
```

```python
econ_activity
```

```python
# Dropping specified columns
econ_activity = econ_activity.drop(columns=['source.label', 'obs_status.label', 'note_classif.label', 'note_indicator.label', 'note_source.label'])
```

```python
# Filter for rows where sex.label is "Sex: Total"
econ_activity = econ_activity[econ_activity['sex.label'] == 'Sex: Total']

# Drop the sex.label column
econ_activity = econ_activity.drop(columns=['sex.label'])

# Verify the changes
print(econ_activity.head())
```

```python
# Filter for rows where classif1.label is "Age (Aggregate bands): Total"
econ_activity = econ_activity[econ_activity['classif1.label'] == 'Age (Aggregate bands): Total']

# Drop the classif1.label column
econ_activity = econ_activity.drop(columns=['classif1.label'])

econ_activity
```

```python
# Filter for ISIC Rev.4 classification
econ_activity_isic = econ_activity[econ_activity['classif2.label'].str.contains('ISIC-Rev.4', na=False)]

# Display the resulting DataFrame
econ_activity_isic.head()
```

```python
# Rename column
econ_activity_isic.rename(columns={'classif2.label': 'sector'}, inplace=True)

# Drop rows with 'Total' in the sector column
econ_activity_isic = econ_activity_isic[~econ_activity_isic['sector'].str.contains('Total', na=False)]

# Remove the prefix "Economic activity (ISIC-Rev.4): " from the sector names
econ_activity_isic['sector'] = econ_activity_isic['sector'].str.replace(r'^Economic activity \(ISIC-Rev\.4\): ', '', regex=True)

# Display the resulting DataFrame
econ_activity_isic.head()
```

```python
# Remove the alphabetical prefixes before the sector names
econ_activity_isic['sector'] = econ_activity_isic['sector'].str.replace(r'^[A-Z]\.\s+', '', regex=True)

# Display the resulting DataFrame
econ_activity_isic.head()
```

```python
# Drop the indicator.label column
econ_activity_isic.drop(columns=['indicator.label'], inplace=True)

# Rename columns
econ_activity_isic.rename(columns={'ref_area.label': 'Country', 'obs_value': 'total employment'}, inplace=True)

# Display the resulting DataFrame
econ_activity_isic.head()
```

```python
# Calculate total employment for each country and year
econ_activity_isic['total_country_year_employment'] = econ_activity_isic.groupby(['Country', 'time'])['total employment'].transform('sum')

# Calculate the share of employment for each sector
econ_activity_isic['share of employment'] = econ_activity_isic['total employment'] / econ_activity_isic['total_country_year_employment']

# Drop the temporary column if it's no longer needed
econ_activity_isic.drop(columns=['total_country_year_employment'], inplace=True)

# Display the resulting DataFrame
econ_activity_isic.head()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Initialize directed graph
G = nx.DiGraph()

# Define the sectors, activities, and capabilities
sectors = {
    "Agriculture & Extractive": ["Planting and harvesting", "Livestock management", "Forestry and logging"],
    "Industrial & Manufacturing": ["Assembly line work", "Quality control", "Machine operation"],
    "Basic Services": ["Customer interaction", "Sales processing", "Stocking shelves"],
    "High-Skill & High-Tech": ["Software development", "Data analysis", "Technical support"],
    "Public & Social": ["Medical care", "Teaching and education", "Public administration"],
    "Creative & Recreational": ["Art creation", "Event management", "Coaching and instructing"],
    "Admin & Info/Management": ["Routine clerical tasks", "Financial advisory", "Information processing"]
}

capabilities = {
    "Planting and harvesting": "Physical capabilities",
    "Livestock management": "Sensory perception",
    "Forestry and logging": "Physical capabilities",
    "Assembly line work": "Physical capabilities",
    "Quality control": "Cognitive capabilities",
    "Machine operation": "Sensory perception",
    "Customer interaction": "Social/emotional capabilities",
    "Sales processing": "Cognitive capabilities",
    "Stocking shelves": "Physical capabilities",
    "Software development": "Cognitive capabilities",
    "Data analysis": "Cognitive capabilities",
    "Technical support": "Social/emotional capabilities",
    "Medical care": "Physical capabilities",
    "Teaching and education": "Social/emotional capabilities",
    "Public administration": "Cognitive capabilities",
    "Art creation": "Cognitive capabilities",
    "Event management": "Cognitive capabilities",
    "Coaching and instructing": "Social/emotional capabilities",
    "Routine clerical tasks": "Cognitive capabilities",
    "Financial advisory": "Cognitive capabilities",
    "Information processing": "Cognitive capabilities"
}

# Add nodes and edges for sectors to activities
for sector, activities in sectors.items():
    for activity in activities:
        G.add_edge(sector, activity)

# Add nodes and edges for activities to capabilities
for activity, capability in capabilities.items():
    G.add_edge(activity, capability)

# Define custom positions for a layered layout
pos = {}

# Set positions for the sectors layer
y_pos = 2
for i, sector in enumerate(sectors.keys()):
    pos[sector] = (i, y_pos)

# Set positions for the activities layer
y_pos = 1
activity_nodes = list(capabilities.keys())
for i, activity in enumerate(activity_nodes):
    pos[activity] = (i % len(sectors), y_pos - (i // len(sectors)) * 0.1)  # Slight shift for clarity

# Set positions for the capabilities layer
y_pos = 0
for i, capability in enumerate(set(capabilities.values())):
    pos[capability] = (i, y_pos)

# Draw the directed graph with the specified positions
plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=9, font_weight="bold", arrows=True)
plt.title("Automation Potential by Sector: Activities and Capability Requirements (Layered)")
plt.show()
```

```python
capability_scores = {
    "Physical Capabilities": 0.7,
    "Cognitive Capabilities": 0.5,
    "Social and Emotional Capabilities": 0.2,
    "Sensory Perception": 0.4
}
```

```python
# Define the capability requirements per ISIC sector, based on the previous breakdown
sector_capabilities = {
    "A - Agriculture, Forestry, and Fishing": ["Physical Capabilities", "Sensory Perception", "Cognitive Capabilities"],
    "B - Mining and Quarrying": ["Physical Capabilities", "Sensory Perception", "Cognitive Capabilities"],
    "C - Manufacturing": ["Physical Capabilities", "Cognitive Capabilities", "Sensory Perception"],
    "D - Electricity, Gas, Steam, and Air Conditioning Supply": ["Physical Capabilities", "Sensory Perception", "Cognitive Capabilities"],
    "E - Water Supply; Sewerage, Waste Management, and Remediation Activities": ["Physical Capabilities", "Cognitive Capabilities", "Sensory Perception"],
    "F - Construction": ["Physical Capabilities", "Cognitive Capabilities", "Sensory Perception"],
    "G - Wholesale and Retail Trade; Repair of Motor Vehicles and Motorcycles": ["Cognitive Capabilities", "Physical Capabilities", "Social and Emotional Capabilities"],
    "H - Transportation and Storage": ["Physical Capabilities", "Cognitive Capabilities", "Sensory Perception"],
    "I - Accommodation and Food Service Activities": ["Social and Emotional Capabilities", "Physical Capabilities", "Cognitive Capabilities"],
    "J - Information and Communication": ["Cognitive Capabilities", "Social and Emotional Capabilities", "Physical Capabilities"],
    "K - Financial and Insurance Activities": ["Cognitive Capabilities", "Social and Emotional Capabilities", "Physical Capabilities"],
    "L - Real Estate Activities": ["Cognitive Capabilities", "Social and Emotional Capabilities", "Physical Capabilities"],
    "M - Professional, Scientific, and Technical Activities": ["Cognitive Capabilities", "Social and Emotional Capabilities", "Physical Capabilities"],
    "N - Administrative and Support Service Activities": ["Cognitive Capabilities", "Social and Emotional Capabilities", "Physical Capabilities"],
    "O - Public Administration and Defense; Compulsory Social Security": ["Cognitive Capabilities", "Social and Emotional Capabilities", "Physical Capabilities"],
    "P - Education": ["Cognitive Capabilities", "Social and Emotional Capabilities", "Physical Capabilities"],
    "Q - Human Health and Social Work Activities": ["Physical Capabilities", "Cognitive Capabilities", "Social and Emotional Capabilities"],
    "R - Arts, Entertainment, and Recreation": ["Physical Capabilities", "Cognitive Capabilities", "Social and Emotional Capabilities"],
    "S - Other Service Activities": ["Physical Capabilities", "Social and Emotional Capabilities", "Cognitive Capabilities"]
}

# Calculate the automation potential for each sector
sector_automation_potential = {}

for sector, capabilities in sector_capabilities.items():
    # Calculate the average automation potential based on capability scores
    total_potential = sum(capability_scores[capability] for capability in capabilities)
    average_potential = total_potential / len(capabilities)
    sector_automation_potential[sector] = average_potential

# Display the automation potential for each sector
sector_automation_potential
```
