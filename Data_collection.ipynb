```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the hierarchical data structure
data_structure = {
    "Agricultural and Extractive Industries": {
        "Activities": [
            "Planting and harvesting",
            "Mining and resource extraction",
            "Basic machinery operation"
        ],
        "Capabilities": [
            "Physical capabilities",
            "Cognitive capabilities",
            "Sensory perception"
        ]
    },
    "Industrial and Manufacturing": {
        "Activities": [
            "Assembly line work",
            "Machine operation",
            "Quality control"
        ],
        "Capabilities": [
            "Physical capabilities",
            "Cognitive capabilities",
            "Sensory perception"
        ]
    },
    "Basic Services": {
        "Activities": [
            "Customer interaction",
            "Logistics and transportation",
            "Routine clerical tasks"
        ],
        "Capabilities": [
            "Social/emotional capabilities",
            "Physical capabilities",
            "Cognitive capabilities"
        ]
    },
    "High-Skill and High-Tech Industries": {
        "Activities": [
            "Data analysis",
            "Financial advisory",
            "Software development"
        ],
        "Capabilities": [
            "Cognitive capabilities",
            "Natural language processing",
            "Social/emotional capabilities"
        ]
    },
    "Public and Social Services": {
        "Activities": [
            "Teaching and education",
            "Medical care",
            "Public administration"
        ],
        "Capabilities": [
            "Social/emotional capabilities",
            "Cognitive capabilities",
            "Physical capabilities"
        ]
    },
    "Creative and Recreational Services": {
        "Activities": [
            "Art creation",
            "Leisure and entertainment services",
            "Personal services"
        ],
        "Capabilities": [
            "Social/emotional capabilities",
            "Physical capabilities",
            "Cognitive capabilities"
        ]
    },
    "Administrative and Information/Management Sector": {
        "Activities": [
            "Miscellaneous administrative tasks",
            "Information processing"
        ],
        "Capabilities": [
            "Cognitive capabilities",
            "Social/emotional capabilities"
        ]
    }
}

# Create a directed graph to represent the hierarchy
G = nx.DiGraph()

# Add nodes and edges for sectors, activities, and capabilities
for sector, details in data_structure.items():
    G.add_node(sector, level=0)
    for activity in details["Activities"]:
        G.add_node(activity, level=1)
        G.add_edge(sector, activity)
    for capability in details["Capabilities"]:
        G.add_node(capability, level=2)
        for activity in details["Activities"]:
            G.add_edge(activity, capability)

# Define layout
pos = nx.multipartite_layout(G, subset_key="level")

# Draw the graph
plt.figure(figsize=(15, 10))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", edge_color="gray", arrows=False)
plt.title("Automation Potential by Sector: Activities and Capability Requirements")
plt.show()
```

```python
# Drop the 'Estimated Automation Potential (%)' column
data_copy = merged_data.drop(columns=['Estimated Automation Potential (%)'])

# Pivot the data to have 'classif2.label' sectors as columns with 'obs_value' as values
pivot_data = data_copy.pivot_table(index=['country', 'year'], columns='classif2.label', values='obs_value')

# Reset the index to make 'country' and 'year' regular columns again
pivot_data.reset_index(inplace=True)

# Display the resulting DataFrame
pivot_data
```

```python
# Sort the data by 'country' and 'year' in descending order
pivot_data.sort_values(by=['country', 'year'], ascending=[True, False], inplace=True)

# Drop duplicate entries for each country, keeping only the first (latest) year
latest_data = pivot_data.drop_duplicates(subset=['country'], keep='first')

# Reset the index for a cleaner look
latest_data.reset_index(drop=True, inplace=True)

# Display the resulting DataFrame
latest_data
```

```python
# Calculate total employment by summing across all sector columns
latest_data['Total Employment'] = latest_data.iloc[:, 2:].sum(axis=1)

# Create percentage columns for each sector
percentage_columns = []
for col in latest_data.columns[2:-1]:  # Exclude 'country', 'year', and 'Total Employment'
    percentage_col = col + ' (%)'
    latest_data[percentage_col] = (latest_data[col] / latest_data['Total Employment']) * 100
    percentage_columns.append(percentage_col)

# Drop the absolute employment columns (all original sector columns)
latest_data = latest_data[['country', 'year'] + percentage_columns]

# Display the resulting DataFrame
latest_data.head()
```

```python
# Define automation potential for each sector based on prior analysis
automation_potentials = {
    'A. Agriculture; forestry and fishing (%)': 60,
    'B. Mining and quarrying (%)': 60,
    'C. Manufacturing (%)': 80,
    'F. Construction (%)': 70,
    'G. Wholesale and retail trade; repair of motor vehicles and motorcycles (%)': 50,
    'H. Transportation and storage (%)': 55,
    'I. Accommodation and food service activities (%)': 45,
    'J. Information and communication (%)': 60,
    'K. Financial and insurance activities (%)': 40,
    'L. Real estate activities (%)': 40,
    'M. Professional, scientific and technical activities (%)': 30,
    'N. Administrative and support service activities (%)': 45,
    'O. Public administration and defence; compulsory social security (%)': 35,
    'P. Education (%)': 20,
    'Q. Human health and social work activities (%)': 25,
    'R. Arts, entertainment and recreation (%)': 15,
    'S. Other service activities (%)': 15
}

# Recalculate weighted automation potential for each country, ignoring missing sector values
latest_data['Automation Potential (%)'] = latest_data.apply(
    lambda row: sum(row[sector] * automation_potentials.get(sector, 0) / 100
                    for sector in automation_potentials.keys() if pd.notnull(row[sector])), axis=1)

# Display the resulting DataFrame to check for any NaN issues
latest_data[['country', 'year', 'Automation Potential (%)']].head() #WE NEED THIS TO COMPARE WITH AUTOMATTION INDEX

```

```python
import matplotlib.pyplot as plt

# Drop any rows with NaN values in the 'Automation Potential (%)' column
latest_data = latest_data.dropna(subset=['Automation Potential (%)'])

# Sort the data by 'Automation Potential (%)' to get the top and bottom countries
top_10_countries = latest_data.nlargest(10, 'Automation Potential (%)')
bottom_10_countries = latest_data.nsmallest(10, 'Automation Potential (%)')

# Plot the top 10 and bottom 10 countries side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 8), sharey=True)
fig.suptitle('Top 10 and Bottom 10 Countries by Automation Potential', fontsize=16)

# Top 10 countries
axes[0].barh(top_10_countries['country'], top_10_countries['Automation Potential (%)'], color='skyblue')
axes[0].set_title('Top 10 Countries with Highest Automation Potential')
axes[0].invert_yaxis()
axes[0].set_xlabel('Automation Potential (%)')

# Bottom 10 countries
axes[1].barh(bottom_10_countries['country'], bottom_10_countries['Automation Potential (%)'], color='salmon')
axes[1].set_title('Bottom 10 Countries with Lowest Automation Potential')
axes[1].set_xlabel('Automation Potential (%)')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
